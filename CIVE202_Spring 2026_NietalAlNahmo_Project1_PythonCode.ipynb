# Load Data

# Code Cell

```python
air_data.columns = air_data.columns.str.strip().str.lower().str.replace('.', '_')
air_data.head()

# Convert date column

air_data['date'] = pd.to_datetime(air_data['date'])

# Clean working dataset

air_clean = air_data.dropna(subset = ['pm2_5_atm', 'pm10_0_atm'])
air_clean.to_csv('air_clean.csv', index = False)
```

# Verify

```python
air_clean.shape

# Missing values (if any)

air_clean.isna().sum()

# Save data

air_clean.to_csv('AirQuality_Daily_StudentVersion (1).csv', index = False)
air_clean.shape
```

# Analysis

```python
# filter PM2.5 >= 5

pm25_filtered = air_clean.query('pm2_5_atm >= 5')
pm25_filtered.shape

# filter PM10 < 6 

pm10_filtered = air_clean.query('pm10_0_atm <6')
pm10_filtered.shape


# Group - monitor_index

by_monitor = air_clean.groupby("monitor_index").agg(
    pm25_mean=("pm2_5_atm", "mean"),
    pm10_mean=("pm10_0_atm", "mean"),
    voc_mean=("voc", "mean"),
    temp_mean=("temperature", "mean"),
    humidity_mean=("humidity", "mean"),
    n=("monitor_index", "size")
).reset_index()

by_monitor.head()
```
